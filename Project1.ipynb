```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency, linregress
from google.colab import files

# Load data
uploaded = files.upload()
titanic_data = pd.read_csv('titanic_train.csv')

# Set Seaborn style for visualizations
sns.set(style='whitegrid')

# 1. Histogram of Age
plt.figure(figsize=(10, 6))
sns.histplot(titanic_data['Age'].dropna(), bins=30, kde=True)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# 2. Pie Chart for Survived
survival_counts = titanic_data['Survived'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(survival_counts, labels=['Not Survived', 'Survived'], autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff'])
plt.title('Survival Distribution')
plt.show()

# 3. Scatter Plot with Regression Line for Age vs. Fare
# Drop rows with missing values in Age or Fare
age_fare_data = titanic_data[['Age', 'Fare']].dropna()
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Fare', data=age_fare_data)
slope, intercept, r_value, p_value, std_err = linregress(age_fare_data['Age'], age_fare_data['Fare'])
plt.plot(age_fare_data['Age'], intercept + slope * age_fare_data['Age'], color='red',
         label=f'Y={slope:.2f}X+{intercept:.2f}\n$R^2$={r_value**2:.2f}')
plt.title('Fare vs Age with Regression Line')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend()
plt.show()

# 4. Additional Figure: Survival Rate by Passenger Class
plt.figure(figsize=(8, 6))
sns.barplot(x='Pclass', y='Survived', data=titanic_data, ci=None)
plt.title('Survival Rate by Passenger Class')
plt.ylabel('Survival Rate')
plt.show()

# 5. Contingency Table and Chi-Square Test for Sex and Survived
contingency_table = pd.crosstab(titanic_data['Sex'], titanic_data['Survived'])
chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f"Chi-squared Statistic: {chi2}")
print(f"P-value: {p}")
print("Contingency Table:\n", contingency_table)

# 6. Comparison of Fare by Embarked Port
plt.figure(figsize=(10, 6))
sns.boxplot(x='Embarked', y='Fare', data=titanic_data)
plt.title('Fare Distribution by Embarked Port')
plt.xlabel('Embarked Port')
plt.ylabel('Fare')
plt.show()

# 7. Fare Distribution by Passenger Class
plt.figure(figsize=(10, 6))
sns.boxplot(x='Pclass', y='Fare', data=titanic_data)
plt.title('Fare Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()

# 8. Comparison of Age and Fare
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Fare', data=titanic_data)
plt.title('Scatter Plot of Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()

# 9. Correlation Matrix of Numeric Columns Only
numeric_data = titanic_data.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_data.corr()

# Plot the correlation matrix
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
